## **GÉOLOCALISATION DES POINTS DE VENTES**

## **Préparation de l'environnement**

### **Chargement des bibliothèques**

```python
#Chargement des librairies Pandas, Network et folium
import pandas as pd
#import networkx as nx
import matplotlib.pyplot as plt
import csv
import folium
import numpy as np
from math import sin, cos, sqrt, atan2, radians

%matplotlib inline
```

### **Importation des données**

```python
#Utilisation de Panda pour la lecture du fichier csv transport_nodes
transport_nodes=pd.read_csv("https://raw.githubusercontent.com/louisisaacdiouf/xyz/master/PDV.csv")
transport_nodes.head()
```

### **Création de la carte**

```python
#Représentation des noeuds sur une carte avec Folium
#La librairie de visualisation Folium a été chargée en haut
#On met en place d'abord le fond de la carte

basemap = folium.Map(
    location=[14.716677, -17.467686],
    zoom_start=12,
    tiles='Open street map'
    )

for index, row in transport_nodes.iterrows():
    folium.Marker([row['latitude'], row['longitude']], 
                  tooltip = 'Lieu : '+str(row['id']),
                  icon=folium.Icon(color='blue', icon='info-sign')).add_to(basemap)
```
